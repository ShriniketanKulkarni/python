```python
#pca
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt, pandas as pd, numpy as np

# Load iris dataset
df=pd.read_csv("Iris.csv")
X=df.iloc[:,[1,2]].values
y=df.iloc[:,-1].values
# Apply PCA to reduce to 2 dimensions
pca = PCA(2)
X_pca = pca.fit_transform(X)

print("Original shape:", X)
print("Reduced shape:", X_pca)
y=pd.factorize(y)[0]
# Plot the 2D projection
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.title("PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
#knn
# knn with confusion matrix
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Load and inspect the dataset
df = pd.read_csv("Iris.csv")
print(df.info())
print(df.head())

# Feature and target extraction
X = df.iloc[:, [1, 2]].values  # SepalLength and SepalWidth
y = df.iloc[:, -1].values      # Species

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train k-NN classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Accuracy
print("Accuracy:", knn.score(X_test, y_test))

# Predict and compute confusion matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot()
plt.title("Confusion Matrix")
plt.show()

# Factorize for color plotting
y_color = pd.factorize(y)[0]

# Scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y_color)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("k-NN Classification")
plt.grid(True)
plt.show()
```

```python
#lwr
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


# Locally Weighted Regression (LWR) implementation

def lwr(X, Y, tau, x):
    w = np.exp(-((X - x)**2).sum(axis=1) / (2 * tau**2))
    Xb = np.c_[np.ones(len(X)), X]
    xb = np.array([1, x[0]])
    W = np.diag(w)
    theta = np.linalg.pinv(Xb.T @ W @ Xb) @ Xb.T @ W @ Y
    return xb @ theta
# Predict for all test points
db = pd.read_csv("housing.csv")
X=db[["housing_median_age"]].values
y=db["median_house_value"].values
xq=np.array([10,20,30,40,50])
y_pred = [lwr(X, y,0.5, np.array([p])) for p in xq]

# Plot results
plt.scatter(X, y, color='b')
plt.scatter(xq, y_pred, color='g')
plt.title('Locally Weighted Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#lr

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
# Load data
df = pd.read_csv('testdata.csv')
X = df.iloc[:, [1]].values # Temperature
y = df.iloc[:, 2].values # Pressure
# Train model
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)
# Plot
plt.scatter(X, y, color='red')
plt.scatter(X, predictions, color='blue')
plt.title('Temperature vs Pressure')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()

print(f"Slope: {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
print(f"MAE: {mean_absolute_error(y, predictions):.4f}")
```

```python
#id3
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix,ConfusionMatrixDisplay

# Load and preprocess data
df = pd.read_csv('Iris.csv')
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
print(df.head())
print(df.info())

# Split data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Train model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.title("Confusion Matrix")
plt.show()


# Predict and evaluate
pred = model.predict(X_test)
acc = accuracy_score(y_test, pred) * 100

print("Predicted labels:", pred.tolist())
print(f"Accuracy: {acc:.2f}%")
```

```python
#naive_bayes
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
# Load dataset
data = load_breast_cancer()
X, y = data.data, data.target

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train naive_bayes classifier
model = GaussianNB()
model.fit(X_train, y_train)

# Test accuracy
accuracy = model.score(X_test, y_test)
print("Test accuracy:", accuracy)
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.title("Confusion Matrix")
plt.show()
```

```python
#ann
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report 

# Load data
df = pd.read_csv("Iris.csv")
X = df.iloc[:, [1, 2]].values  # Sepal Length and Width
y = df.iloc[:, -1]  # Encode species to 0/1/2

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train ANN
model = MLPClassifier(max_iter=1000, random_state=1)
model.fit(X_train, y_train)

# Evaluate
print("Accuracy:", model.score(X_test, y_test))
print(classification_report(y_test, model.predict(X_test)))
y=pd.factorize(y)[0]

#confusion_matrix
accuracy = model.score(X_test, y_test)
print("Test accuracy:", accuracy)
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
# Plot
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("ANN Classification")
plt.grid(True)
plt.show()

```
